## 4. Задание (в программе):


Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
```

```python
k1 = 1
k2 = 3
```

```python
y1 = np.cos(k1 * x)
y2 = np.cos(k2 * x)
```

```python
plt.plot(x, y1, color="green", label="k=1")
plt.plot(x, y2, color="blue", label="k=3")
plt.title('y(k,x)=cos(k∙x)')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
