2.	Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import math 
```

### 3. Задание (в программе)

Напишите код на Python, реализующий построение графиков:

эллипса,
гиперболы.


```python
#окружность
```

```python
R = 5
x = np.linspace(-R,R,100)
y1 = np.sqrt(R**2 - x**2)
y2 = -np.sqrt(R**2 - x**2)
```

```python
plt.plot(x, y1, color="green",)
plt.plot(x, y2, color="green",)
plt.title('окружность')

plt.axis('equal')
plt.axis([-R, R, -R, R])
plt.grid()
plt.show()
```

```python
#элипс
# x^2/a^2 + y^2/b^2 = 1
```

```python
a = 3
b = 2
x = np.linspace(-a,a,100)
y1 = np.sqrt((1 - x**2/a**2)*b**2)
y2 = -np.sqrt((1 - x**2/a**2)*b**2)
```

```python
plt.plot(x, y1, color="green",)
plt.plot(x, y2, color="green",)
plt.title('эллипс')

plt.axis('equal')
plt.axis([-4, 4, -4, 4])
plt.grid()
plt.show()
```

```python
#гипербола
# y = 1/x
```

```python
x1 = np.linspace(0.01,10,100)
y1 = 1/x1
x2 = np.linspace(-10,-0.01,100)
y2 = 1/x2
```

```python
plt.plot(x1, y1, color="green",)
plt.plot(x2, y2, color="green",)
plt.title('гипербола')

plt.axis('equal')
plt.axis([-6, 6, -6, 6])
plt.grid()
plt.show()
```

### 4. Задание (в программе)

1. Решите систему уравнений:
y = x^2 – 1;
exp(x) + x∙(1 – y) = 1


```python
from scipy.optimize import fsolve
```

```python
def equations(p):
    x,y = p
    return (y - x**2 + 1, np.exp(x) + x*(1-y) - 1)
```

```python
x1,y1 = fsolve(equations, (3,3))
x1,y1
```

2. Решите систему уравнений и неравенств:
y = x^2 – 1;
exp(x) + x∙(1 – y) > 1


```python
x = np.linspace(-10,10,100)
y1 = x**2 - 1

# exp(x) + x*(1 – y) > 1
# x*(1 – y) > 1 - exp(x)
# 1 – y > (1 - exp(x))/x
# y < 1 - (1 - exp(x))/x

y2 = 1 - (1 - np.exp(x))/ x

plt.plot(x, y1)
plt.plot(x, y2, color="red",)


plt.axis([-5, 5, -2, 20])

plt.grid()
plt.show()
```

```python
# перечесение справа
x1,y1 = fsolve(equations, (3,6))
x1,y1
```

```python
# перечесение слева
x2,y2 = fsolve(equations, (-2,1))
x2,y2
```
